```python
def count_industry_groups(data, column_name):
    industry_groups = data[column_name].astype(str).str.split(',')  
    all_groups = [group.strip() for groups in industry_groups for group in groups]  

    group_counts = pd.Series(all_groups).value_counts()  

    result_df = pd.DataFrame({'Industry': group_counts.index, 'Count': group_counts.values})

    return result_df


data1 = data
column_name = 'Industries'  
a1 = count_industry_groups(data1, column_name)

#
print(a1)
```

```python
def count_industry_funding_round(data):
    industry_counts = {}

    for index, row in data.iterrows():
        industries = str(row['Industries']).split(', ')
        funding_rounds = row['Number of Funding Rounds']

        for industry in industries:
            if industry not in industry_counts:
                industry_counts[industry] = funding_rounds
            else:
                industry_counts[industry] += funding_rounds

    result_df = pd.DataFrame({'Industry': list(industry_counts.keys()), 'Funding Count': list(industry_counts.values())})
    return result_df

# Örnek kullanım
data4 = data[['Industries','Number of Funding Rounds']]

a2 = count_industry_funding_round(data4)
print(a2)
```

```python
def count_industry_funding2(data):
    industry_counts = {}

    for index, row in data.iterrows():
        industries = str(row['Industries']).split(', ')
        funding_rounds = row['Total Funding Amount Currency (in USD)']

        for industry in industries:
            if industry not in industry_counts:
                industry_counts[industry] = funding_rounds
            else:
                industry_counts[industry] += funding_rounds

    result_df = pd.DataFrame({'Industry': list(industry_counts.keys()), 'Funding Total': list(industry_counts.values())})
    return result_df

# Örnek kullanım
data5 = data[['Industries','Total Funding Amount Currency (in USD)']]

a3 = count_industry_funding2(data5)
print(a3)
```

```python
merged_df = pd.merge(a1, a2, on='Industry')
merged_df = pd.merge(merged_df, a3, on='Industry')
```
